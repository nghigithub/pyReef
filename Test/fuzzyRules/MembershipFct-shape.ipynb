# Examples of membership function shapes

We use **scikit-fuzzy** (a.k.a. _skfuzzy_): Fuzzy Logic Toolbox for Python to define our fuzzy logic carbonate rules. [SkFuzzy](http://pythonhosted.org/scikit-fuzzy/) contains a series of built-in membership functions that can be combined to define your own rules. Here we present some of the most used ones for PyReef.

For a complete description, the reader/user could have a look at the complete API for membership generation available [here](http://pythonhosted.org/scikit-fuzzy/_modules/skfuzzy/membership/generatemf.html#sigmf).

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'
```

```python
def plot_fuzzy_curve(bathy,law,label,color='b'):
    fig, ax0 = plt.subplots(figsize=(8, 4))
    ax0.plot(bathy, law, color, linewidth=4, label=label)
    lg = ax0.legend(loc=1)
    lg.draw_frame(False)
    plt.grid(b=True, which='both', color='0.65',linestyle='-')
    plt.xlabel('X-axis')
    plt.ylabel('Fuzzy value')
    plt.show()
    return
```

```python
x = np.linspace(0.0, 50.0, num=201, endpoint=True)
```

# Membership function triangular

```python
tri = fuzz.trimf(x, [10, 20, 45])

# Visualise fuzzy production curve
plot_fuzzy_curve(x,tri,'triangular shape','k')
```

# Membership function trapezoidal

```python
a = 5
b = 10
c = 25
d = 40
trap = fuzz.trapmf(x, [a, b, c, d])

# Visualise fuzzy production curve
plot_fuzzy_curve(x,trap,'Trapezoidal shape','k')
```

# Membership function gaussian

[Definition](https://ned.ipac.caltech.edu/level5/Leo/Stats2_3.html)

```python
gaussian_sigma = 10
gauss = fuzz.gaussmf(x, 25, gaussian_sigma)

# Visualise fuzzy production curve
plot_fuzzy_curve(x,gauss,'gaussian shape','k')
```

# Membership function bell shape

[Definition of parameters a,b & c](http://www.bindichen.co.uk/post/Fundamentals/bell-shaped-function.html)

```python
a = 10.
b = 5
c = 20.
bell = fuzz.gbellmf(x, a, b, c)

# Visualise fuzzy production curve
plot_fuzzy_curve(x,bell,'bell shape','k')
```
